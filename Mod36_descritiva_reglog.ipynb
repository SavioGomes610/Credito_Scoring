## Clculo do *WOE*

**Lembrete**:

O WOE (*weight of evidence*)  uma medida de relevncia das categorias de uma varivel para classificao binria, alinhada com a regresso logstica. O *WOE*  uma medida comparativa que indica o quanto um perfil  diferente da populao geral em termos da probabilidade do evento que estamos estudando. Uma forma interessante de se entender o *WOE*  como o logaritmo da razo de chances de ocorrncia do evento no perfil observado com relao  chance geral da populao. Abaixo uma tica matemtica sobre como calcular este valor para a categoria *i* ($WOE_i$):

$$ WOE_i=log(RC_i) $$

em que $RC_i$  a razo de chances do perfil $i$ com relao ao pblico geral. Ou seja:

$$ WOE_i=log(RC_i)=log(\frac{Odds_i}{Odds_{geral}}) $$
ou ainda:
$$ WOE=log(Odds_i) - log(Odds_{geral}) $$

Lembrando que $Odds_i$  o nmero de eventos do perfil i dividido pelo nmero de no eventos no mesmo perfil. Que tambm pode ser calculado como $P_i/(1-P_i)$.

```python
tab = pd.crosstab(ttnc.sex, ttnc.survived, margins=True, margins_name='total')
tab
```

```python
tab['bad_rate'] = tab[1]/tab['total']
tab
```

### Forma prtica de clculo

Na forma acima, precisamos fazer muitas operaes para chegar no WOE. Abaixo vamos descrever uma forma mais prtica.

Essa razo $Odds_i/Odds_{geral}$  a razo de chances (odds ratio) do perfil i sobre o perfil geral. Considere a seguinte tabela cruzada:

|Perfil|Eventos|No eventos|Total|
|:|:|:|:|
|1|$a_{11}$|$a_{12}$|$a_{1.}$|
|2|$a_{21}$|$a_{22}$|$a_{2.}$|
|Total|$a_{.1}$|$a_{.2}$|$a_{..}$|

Uma forma prtica de calcular esse *odds ratio*  a seguinte:

- Calculamos as propores para as categorias da varivel (dado evento, e dado no evento): 
$$\%eventos_i = \frac{a_{i1}}{a_{.1}}$$
$$\%no\ eventos_i = \frac{a_{i2}}{a_{.2}}$$
<br>
- O odds ratio da categoria i com relao ao perfil geral ser: 
$$OR = \frac{\%eventos_i}{\%no\ eventos_i}$$

de onde obtemos:

$$WOE_i = log\left(\frac{\%eventos_i}{\%no\ eventos_i}\right) $$

```python
tab = pd.crosstab(ttnc.sex, ttnc.survived, margins=True, margins_name='total')
tab
```

```python
tab['pct_nao_evento'] = tab[0]/tab.loc['total',0]
tab['pct_evento']     = tab[1]/tab.loc['total',1]
tab
```

```python
tab['RC'] = tab['pct_evento']/tab['pct_nao_evento']
tab['WOE'] = np.log(tab['pct_evento']/tab['pct_nao_evento'])
tab
```

## *Information Value* (IV)

O *Information Value*  uma medida de poder preditivo muito popular na rea de crdito.  um indicador que mede o quanto uma varivel  capaz de 'separar' individualmente eventos e no eventos, com base no WOE, que est indimamente associado com a funo linear da regresso logstica.

O *IV* pode ser calculado assim:

$$ \sum_{i=1}^{num-classes} \left(\%eventos_i - \%no\ eventos_i\right)WOE_i $$

Naeem Siddiqi recomenda as seguintes regras de bolso para interpretar o IV:

|Faixa|Poder preditivo|
|:-|:-|
|0 a 0,02| Intil |
|0,02 a 0,1| Fraco |
|0,1 a 0,3| Mdio |
|0,3 a 0,5| Forte |
|0,5 ou mais| Suspeito de to alto |

Lembro que Siddiqi recomenda estas regras de bolso da experincia dele, que  muito vasta em modelos de crdito. Outras reas de concentrao podem tender a apresentar relaes mais suts ou mais evidentes.

O IV parte de uma tabela cruzada. Vamos fazer um exemplo com a base do Titanic:

```python
tab = pd.crosstab(ttnc.sex, ttnc.survived, margins=True, margins_name='total')
tab
```

Agora vamos calcular o percentual de eventos e o percentual de no eventos. Feito isto, j vimos que fica bastante fcil calcular o WOE.

```python
rtulo_nao_evento = tab.columns[1]
rtulo_nao_evento
```

```python
tab = pd.crosstab(ttnc.sex, ttnc.survived, margins=True, margins_name='total')

rtulo_evento = tab.columns[0]
rtulo_nao_evento = tab.columns[1]

tab['pct_evento'] = tab[rtulo_evento]/tab.loc['total',rtulo_evento]
tab['pct_nao_evento'] = tab[rtulo_nao_evento]/tab.loc['total',rtulo_nao_evento]
tab['woe'] = np.log(tab.pct_evento/tab.pct_nao_evento)
tab
```

Com as quantidades calculadas na clula acima, fica bem fcil calcular o *Information Value*.

Vamos fazer o clculo das parcelas no *data frame*, mas como veremos na clula seguinte, podemos calcular em uma nica linha:

```python
tab['iv_parcial'] = (tab.pct_evento - tab.pct_nao_evento)*tab.woe
print(tab.iv_parcial.sum())
tab
```

### Calculando o IV para um conjunto grande de variveis

Agora vamos definir uma estratgia para calcular o IV para um conjunto grande de variveis. Essa estratgia vai ser assim:

1. Vamos construir um metadados (*metadata*), que  uma tabela que vai conter informaes sobre as variveis: tipo de dado, nmero de categorias etc, inclusive o *Information Value*.
2. Vamos atribuir papis a essas variveis, como "resposta" - a varivel que vamos prever, "covarivel" as variveis que vo ajudar a prever a resposta
3. Vamos definir uma funo que calcula o IV
4. Vamos calcular o IV para cada varivel tipo "covariavel" e popular o metadados.

**1. Definindo o metadados**

```python
ttnc.drop(columns=['class', 'deck', 'embark_town', 'alive'], inplace=True)
metadados = pd.DataFrame({'dtype':ttnc.dtypes})
metadados
```

**2. Atribuindo papis**

A atribuio de papis pode ser til para localizarmos determinadas variveis em um banco de dados especialmente grande. Em projetos maiores,  muito til manter uma lista das variveis da tabela em excel mesmo, com algumas anotaes sobre elas. Tipicamente, a primeira lista de variveis vem da aplicao que usamos, no caso o Python, e j vem com algumas informaes que calculamos.

Vamos usar esse metadados para identificar as variveis explicativas, ou "covariveis" como estamos chamando aqui (ou ainda *features*, como se diz na literatura *machinelearnica*).

```python
metadados['papel'] = 'covariavel'
metadados.loc['survived','papel'] = 'resposta'
metadados['nunique'] = ttnc.nunique()
metadados
```

**3. Definindo a funo**

Vamos fazer uma funo que recebe como entrada o nome de uma *feature* (ou varivel explicativa), o nome da varivel resposta (ou *target*), e retorna o IV.

```python
def IV(variavel, resposta):
    tab = pd.crosstab(variavel, resposta, margins=True, margins_name='total')

    rtulo_evento = tab.columns[0]
    rtulo_nao_evento = tab.columns[1]

    tab['pct_evento'] = tab[rtulo_evento]/tab.loc['total',rtulo_evento]
    tab['pct_nao_evento'] = tab[rtulo_nao_evento]/tab.loc['total',rtulo_nao_evento]
    tab['woe'] = np.log(tab.pct_evento/tab.pct_nao_evento)
    tab['iv_parcial'] = (tab.pct_evento - tab.pct_nao_evento)*tab.woe
    return tab['iv_parcial'].sum()

iv = IV(ttnc.sex, ttnc.survived)
iv
```

**4. Populando o metadados**

Agora vamos criar uma coluna nova no metadados e popular com o IV da respectiva varivel.

Um dos maiores desafios de se automatizar uma tarefa como essa de calcular o IV para muitas variveis  tratar as excees. Quando h categorias com pouca frequencia, o clculo do IV pode resultar em uma diviso por zero. Isso costuma ocorrer com uma certa frequencia quando a varivel  contnua.

Vamos fazer um tratamento bsico, at um tanto grosseiro, de categorizar a covarivel em 5 grupos caso ela tenha muitas categorias.

Mas pode ser importante fazer um pr-processamento ou uma lgica mais sofisticada dependendo do problema que se estiver encontrando.

```python
for var in metadados[metadados.papel=='covariavel'].index:
    if  (metadados.loc[var, 'nunique']>5):
        metadados.loc[var, 'IV'] = IV(pd.qcut(ttnc[var],5,duplicates='drop'), ttnc.survived)
    else: 
        metadados.loc[var, 'IV'] = IV(ttnc[var], ttnc.survived)
    
metadados
```

# Anlise longitudinal

Uma parte importante da anlise em modelos que se aplicam em grandes empresas - como os modelos de crdito -  a anlise longitudinal.

 esperado que populaes e mercados sofram mudanas o tempo todo. Algumas delas podem ser bruscas, como os chamados cisnes negros (ou *black swams*), outras podem ser suts.  uma tarefa muito difcil fazer previses acuradas de eventos como *black swams* antes que eles aconteam (ex: quebra da bolsa, crise imobiliria, pandemia, etc). Mesmo eventos mais suts tambm podem ser igualmente difceis de se prever - economistas esto o tempo todo tentando prever inflao, cmbio, valores de aes, tudo com sucessos extremamente limitados, e altas doses de risco.

Por esse motivo,  muito comum (e at pragmtico) deixar esse tipo de anlise em uma etapa descritiva e indicadora dos riscos contidos no modelo. Se uma varivel apresenta instabilidade, talvez seja melhor remov-la do modelo.

Esse tipo de processo, como o dos modelos de crdito, que necessita de previses mensais, tem duas caractersticas importantes: o patamar (ou calibragem) e a discriminao. O patamar pode ser muito difcil de ser previsto, em especial quando o cenrio  muito voltil ou quando o horizonte de previso  muito longo. Mas a **razo de chances** entre dois indivduos com classificaes diferentes  mais fcil de se acertar.

## WOE no tempo

Uma anlise clssica para se avaliar se uma varivel  fator de risco *consistente* no tempo  a avaliao do WOE ao longo do tempo. A distncia entre o WOE de duas categorias de uma varivel  uma estimativa emprica do valor do parmetro dessa varivel na regresso logstica. Assim, se os WOEs forem estveis no tempo, a varivel vai ser um preditor consistente e no vai trazer surpresas.

Podemos construir a mesma avaliao nas *taxas de evento observadas*, mas essas quantidades podem ser poludas por mudanas sistmicas na taxa de evento, o que no quer dizer que a *razo de chances* vai ser instvel no tempo.

Vamos ver isso nos dados agregados que contm os visitantes de uma pgina web para recebimento de doaes de uma instituio beneficente, indicando o sexo do visitante, e se virou doador ou no. Esses dados possuem uma certa *sazonalidade*, um perodo em que aumentam as taxas de converso para doador em todos os perfis sistematicamente.

Vamos ver como a taxa de doadores e o WOE se comportam ms a ms

```python
doadores = pd.read_pickle('doadores.pkl')
doadores.head()
```

```python
tab = doadores.unstack()['freq']
tab['N'] = tab.sum(axis=1)
tab['tx_evento'] = tab.sim/tab.N
tab.reset_index()
tab.head()
```

```python
fig, ax = plt.subplots()

sns.pointplot(data=tab.reset_index(), x='data_ref', y='tx_evento', hue='sexo', ci=None, ax=ax)
ax.set_ylim([0, tab.tx_evento.max()*1.2]);

# Configurar rtulos do eixo X
labs = tab.index.get_level_values(level=0).unique()
labs2 = [x.to_period('M') for x in labs]

ax.set_xticklabels(labs2,rotation=30, fontsize='small')
ax.legend(bbox_to_anchor=(.65, -.2), ncol=2)
```

Podemos reparar que o perodo central do grfico acima possui um aumento com relao aos demais, para ambos os perfis.  muito comum que o interesse central do modelo seja simplesmente identificar os perfis propensos a maiores taxas de eventos, o que pode ser avaliado de forma mais precisa utilizando o WOE.

```python
tab.head()
```

```python
rotulo_ev, rotulo_nev = ('sim', 'nao')

tab['pct_evento']     = tab[rotulo_ev]/tab[rotulo_ev].groupby(level=0).sum()
tab['pct_nao_evento'] = tab[rotulo_nev]/tab[rotulo_nev].groupby(level=0).sum()
tab['WOE'] = np.log(tab.pct_evento/tab.pct_nao_evento)
tab.head()
```

```python
fig, ax = plt.subplots()
variavel='sexo'

# Comando principal do grfico
sns.pointplot(data=tab.reset_index(), y = 'WOE', x = 'data_ref', hue=variavel, ax=ax);

# Rotacionar os rtulos do eixo X
labs = tab.index.get_level_values(level=0).unique()
labs2 = [x.to_period('M') for x in labs]

ax.set_xticklabels(labs2,rotation=30, fontsize='small')
ax.legend(bbox_to_anchor=(.65, -.2), ncol=2)

#limites do grfico - mostrando 1 e -1 para ter referncia
ymax = max(1.2, max(tab.WOE))
ymin = min(-1.2, min(tab.WOE))
ax.set_ylim([ymin, ymax])

# Linhas de referncia
ax.axhline(1, linestyle=':'  , color='red', alpha=.5)
ax.axhline(0, linestyle='--' , color='grey', alpha=.5)
ax.axhline(-1, linestyle=':' , color='red', alpha=.5)

# Ttulos
ax.set_title(f'WOE ao longo das safras para varivel {variavel.capitalize()}')
ax.set_xlabel('Safra', loc='right');
```

Como podemos observar, o WOE, que est associado aos parmetros da regresso logstica, no  afetado por perodos de alta sistmica.

Neste caso, vemos que o WOE permanece bastante estvel no tempo, indicando uma varivel bem robusta e confivel.

### Calculando o IV no tempo

Uma faceta importante de muitos projetos reais de cincia de dados est em efeitos temporais que podem ocasionar comportamentos preditivos muito bons, mas inconsistentes no tempo.

Por isso  muito importante realizar anlises no tempo, para ter mais segurana de que uma varivel com alto poder preditivo continue assim quando o modelo for aplicado.

```python
tab.head()
```

```python
((tab.pct_evento - tab.pct_nao_evento)*tab.WOE)
```

```python
iv_sexo = ((tab.pct_evento - tab.pct_nao_evento)*tab.WOE).groupby(level=0).sum();
iv_sexo.plot.bar()
```

```python
def plot_iv(IV):
    fig, ax = plt.subplots()
    x = IV.index
    a = [.02]*len(IV.index)
    b = [.1]*len(IV.index)
    c = [.3]*len(IV.index)
    d = [.5]*len(IV.index)
    e = [.6]*len(IV.index)

    ax.fill_between(IV.index, a, color='grey', alpha=.2)
    ax.fill_between(IV.index, a, b, color='orange', alpha=.1)
    ax.fill_between(IV.index, b, c, color='green', alpha=.1)
    ax.fill_between(IV.index, c, d, color='blue', alpha=.1)
    ax.fill_between(IV.index, d, e, color='purple', alpha=.1)
    ax.plot(IV)

    labs = IV.index.get_level_values(level=0).unique()
    labs2 = [x.to_period('M') for x in labs]

    ax.set_xticks(labs)
    ax.set_xticklabels(labs2,rotation=30, fontsize='small')

    ax.set_title("Information Value de Sexo ao longo das safras")
    ax.set_ylabel("Information Value")
    ax.set_xlabel("Safra")
    
plot_iv(iv_sexo)
```
